```python

# create a list of records
tweet_info_ls = []

# iterate over each tweet and corresponding user details
for tweet, user in zip(tweets.data, tweets.includes['users']):
    tweet_info = {
        'created_at': tweet.created_at,   #date and time of the tweet
        'ID':tweet.id,                    #ID of the post
        'Post': tweet.text,               #text of the tweet
        'Source': tweet.source,           #which device was used for the tweet
        'Screen_Name': user.name,         #screen name of the user 
        'user_Name': user.username,       #user name set in the profile
        'Location': user.location,        #location of the user when s/he shares this tweet
        'Varified': user.verified,        #whether the post is varified or not
        'Bio_User': user.description      #profile description of the user
    }
    tweet_info_ls.append(tweet_info)

# create dataframe from the extracted records
tweets_df = pd.DataFrame(tweet_info_ls)
# display the dataframe
tweets_df.head(20)
```
